## pair grids

```python
iris.columns
```

```python
x = sns.PairGrid(iris , vars = ['sepal_length' , 'petal_length'] ,\
                 hue='species' , palette='husl')
x = x.map_diag(plt.hist , edgecolor= 'white')
x = x.map_offdiag(plt.scatter , edgecolor= 'black')
x = x.add_legend()
```

```python
x = sns.PairGrid(iris , x_vars = ['petal_length' , 'petal_width'], \
                y_vars=['sepal_length' , 'sepal_width'] , \
                hue = 'species')

x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris , hue='species' , palette='coolwarm')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

## Regression Plot

```python
sns.regplot(x= 'total_bill' , y = 'tip' , data=tips)
```

```python
mean = [2,5]
cov = [[1.1,0.4] , [2.2 , 3]]

x_val , y_val = np.random.multivariate_normal(mean , cov , 100).T
```

```python
sns.regplot(x=x_val , y=y_val , color='purple' , marker = 'D')
```

```python
sns.regplot(x= 'total_bill' , y = 'tip' , data=tips , marker= '+',\
           scatter_kws={'color' : 'blue' , 's' : 15},\
           line_kws={'color' : 'red' , 'linewidth' : 3},\
           ci=68)
```

```python
sns.regplot(x = 'size' , y = 'total_bill' , data= tips , x_jitter=0.3)
```

```python
sns.regplot(x = 'size' , y = 'total_bill' , data= tips , \
            x_estimator=np.mean , x_ci=68)
```
